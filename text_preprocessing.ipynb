```python

from transformers import CLIPTokenizer

# Sample text descriptions
descriptions = [
    "A yellow flower with round petals.",
    "A red flower with a long green stem.",
    "Purple flower with wide petals."
]

# Load tokenizer
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")

# Tokenize text
tokens = tokenizer(descriptions, padding=True, truncation=True, return_tensors="pt")
print("Token IDs:\n", tokens["input_ids"])
print("Attention Masks:\n", tokens["attention_mask"])
```
